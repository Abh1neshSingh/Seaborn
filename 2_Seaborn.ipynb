```python
# pie chart plot using matplotlib
```

```python
lang=['c','python','java']
student=[20,100,40]

plt.pie(student,labels=lang)
plt.title('Pie chart')
plt.show()
```

```python
# if you want fraction of this  plot in Percentage 
# then use-- autopct='%1.1f%%'
```

```python
lang=['c','python','java']
student=[20,100,40]

plt.pie(student,labels=lang,autopct='%1.1f%%')
plt.title('Pie chart')
plt.show()
```

```python
# here we  divide the plot into Separate part
```

```python
my_explode= (0.0,1.0,0)

lang=['c','python','java']
student=[20,100,40]

plt.pie(student,labels=lang,autopct='%1.1f%%',explode=my_explode,shadow=True)
plt.title('Pie chart')
plt.show()

```

```python
my_explode= (0.0,0.2,0.2)

lang=['c','python','java']
student=[20,100,40]

plt.pie(student,labels=lang,autopct='%1.1f%%',explode=my_explode,shadow=True)
plt.title('Pie chart')
plt.show()
```
